# Demo: Datmo.io

In this notebook, we'll be going over a valuable tool called Datmo, which will be especially helpful if you decide to pursue data science further. In fact, both of our organizations strive to improve access to critical data-driven tools through education and useful products, respectively. 

Datmo is a *command line interface* that helps make your workflow a lot more efficient. You'll be able to run and keep track of *machine learning* algorithms (somewhat like Github) and share your experiements with others.

We'll be using Datmo on a couple problems from the same dataset as one of DSI's case studies on engineering graduates' salaries. 

In our basic curriculum, we go from teaching basic Python to implementing a basic machine learning algorithm: *linear regression,* where we essentially make a *model* from our dataset based on the equation of a line. In this example, we're going to see if different factors (like test scores) affect someone's salary.

```python
import numpy as np
from datascience import *
from matplotlib import pyplot as plt
import random
%matplotlib inline
plt.style.use('fivethirtyeight')
```

To start, let's read our CSV. Data from: http://research.aspiringminds.com/resources/#ameo

```python
salaries = Table().read_table("engineering_data.csv")
salaries
```

Let's first try to predict salaries from 12th grade scores. We chose the 12th grade scores over 10th grade scores, as they've obviously picked up more relevant skills since then, and over college GPA because that takes into account courses that may not have to do with the subjects' careers.

Our algorithm will only be reliable, however, if all the students take the same exam! Let's clean out the data before we plot it.

```python
salary_12 = salaries.where("12board", are.equal_to("cbse")).select("Salary", "12percentage")
```

```python
salary_12.scatter("12percentage", "Salary", fit_line=True)
```

To calculate the correlation coefficient, we do the following to convert our data into standard units.

```python
#Calculating distance from the mean, dividing by standard deviation
def standard_units(nums):
    return (nums - np.mean(nums))/np.std(nums)
#Average of the product of x and y values in standard units 
def calc_r(tbl, col_1, col_2):
    return np.mean(standard_units(tbl.column(col_1)) * standard_units(tbl.column(col_2)))
```

```python
r = calc_r(salary_12, "12percentage", "Salary")
print("The correlation coefficient between 12th grade CBSE scores and salaries is " + str(r) + ". \nThis indicates a slight positive correlation.")
```

Now, let's calculate the values for our regression line.

```python
def values(tbl, r, col_1, col_2):
    m = r * np.std(tbl.column(col_2)) / np.std(salary_12.column(col_1))
    b = r * np.mean(tbl.column(col_2)) - m * np.mean(tbl.column(col_1))
    return m, b
line_vals = values(salary_12, r, "12percentage", "Salary")
print("Slope: " + str(line_vals[0]))
print("Intercept: " + str(line_vals[1]))
```

The process we just went through is called *training* a model. Note how the first column of our table said "train". Indeed, this is because in machine learning, we often use two sets of data: a *training* set and a *test* set.

```python
test_data = Table().read_table("engineering_salaries_test.csv")
test_data
```

We're going to use the *parameters* defined to predict the salaries of these workers, whom we've never seen before! As we can see, this model underestimates most workers' salaries. One of the reasons for this is that we have workers in the test set whose scores are outside the range of the training set.

```python
def predict_salary(arr, m, b):
    return arr * m + b
salary_arr = predict_salary(test_data.column("12percentage"), line_vals[0], line_vals[1])
plt.scatter(salary_12.column("12percentage"), salary_12.column("Salary"))
plt.scatter(test_data.column('12percentage'), salary_arr)
```

Let's now see how accurate this model is, with half of our first dataset serving as the training set and the other half as our test set.

```python
cbse_engineers = salaries.where("12board", are.equal_to("cbse")).select("12percentage", "Salary")
train = cbse_engineers.take(range(0, cbse_engineers.num_rows // 2))
test = cbse_engineers.take(range(cbse_engineers.num_rows // 2, cbse_engineers.num_rows))

train_r = calc_r(train, "12percentage", "Salary")
train_values = values(train, train_r, "12percentage", "Salary")
train_m = train_values[0]
train_b = train_values[1]

print("r = " + str(train_r))
print("m = " + str(train_m))
print("b = " + str(train_b))
print("We can see that these quantities are close to the ones for the entire dataset.")
predicted_salaries = predict_salary(test.column("12percentage"), train_m, train_b)
```

The array `predicted_salaries` contains our predictions using the model. Now, let's compare these values to the real data. The orange points represent the actual data and the blue points represent our model.

```python
test_comparison = test.with_column("Predictions", predicted_salaries)
plt.scatter(test_comparison.column(0), test_comparison.column(1))
plt.scatter(test_comparison.column(0), test_comparison.column(2))
test_comparison
```

```python
def tse_tbl(tbl):
    def total_squared_error(slope, intercept):
        x_val = tbl.column(0)
        y_val = tbl.column(1)
        predictions = (slope * x_val) + intercept
        errors = (predictions - y_val)**2
        return sum(errors)
    return total_squared_error
print("Total Squared Error: " + str(tse_tbl(test)(train_m, train_b)))
def ave_pct_error(tbl):
    return str((np.mean(np.abs(tbl.column(2) - tbl.column(1)) / tbl.column(1))) * 100) + "%"
print("Average Percent Error: " + ave_pct_error(test_comparison))
```

As we can see, this error is quite large! Let's now try generating our values for the regression line using an algorithm called *least squares*, which seeks to minimize the squared error of our prediction. For the sake of this exercise, you don't need to know how it actually works, but you're welcome to look it up yourself!

The `minimize` function from the `datascience` library outputs the parameters that would minimize its output in the form of an array. 

```python
min_params = minimize(tse_tbl(train))
print("Minimized m: ", min_params[0])
print("Minimized b: ", min_params[1])
print("Previous m: ", train_m)
print("Previous b: ", train_b)
```

Let's now use this model on our test dataset.

```python
print("Total squared error: ", tse_tbl(test)(min_params[0], min_params[1]))
print("Previous total squared error: ", str(tse_tbl(test)(train_m, train_b)))
```

```python
ls_comparison = test.with_columns("Predictions", 
                                  predict_salary(test.column("12percentage"), min_params[0], min_params[1]))
ls_comparison
```

```python
print("Average Percent Error: ", ave_pct_error(ls_comparison))
print("Previous Average Percent Error: ", ave_pct_error(test_comparison))
```

By using least-squares, we were able to reduce our percent error, on average, by over 20%! Let's now plot our data.

```python
plt.scatter(ls_comparison.column(0), ls_comparison.column(1))
plt.scatter(ls_comparison.column(0), ls_comparison.column(2))
```
